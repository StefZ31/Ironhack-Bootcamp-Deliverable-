```python
#unique values from bedrooms, bathrooms, floors, condition, grade
ubedrooms=house_price_regression.bedrooms.nunique() #bedrooms
ubathrooms=house_price_regression.bathrooms.nunique() #bathrooms
ufloors=house_price_regression.floors.nunique() #floors
ucondition=house_price_regression.condition.nunique() #condition
ugrade=house_price_regression.grade.nunique() #grade
print ('These are the unique values of each column:')
print ('Bedrooms:', ubedrooms)
print ('Bathrooms:', ubathrooms)
print ('Floors:', ufloors)
print ('Condition:', ucondition)
print ('Grade:', ugrade)
```

```python
#we use mean to get the avg value for a specific column
avgprice=house_price_regression['price'].mean()
print('The average price for all the properties is:',round((avgprice),0),'(Currency: EUR, USD, or any other)')
```

```python
#groupby avg price houses grouped bedrooms, column bedrooms and avg price
bp=house_price_regression.groupby('bedrooms')['price'].mean()
bp.round(0).astype(int) #changed price to int type, otherwise it returns a float value which looks really wrongly formated
```

```python
#groupby avg sqft living grouped by bedrooms, column bedrooms and avg sqft living
bavsql=house_price_regression.groupby('bedrooms')['sqft_living'].mean()
bavsql.round(0).astype(int)
```

```python
#groupby avg prices with waterfront, columns, waterfront and avg price
wp=house_price_regression.groupby('waterfront')['price'].mean()
wp.round(0).astype(int)
```
